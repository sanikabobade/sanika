```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
ds=pd.read_csv('Salary_Data.csv')
ds
```

```python
# divide dataset as independant(input) and dependant(output) variables

x=ds[['YearsExperience']]
y=ds[['Salary']]
```

```python
ds.columns
```

```python
ds.info()
```

```python
x
```

```python
# plot graph(positive corr)

plt.scatter(x,y)
plt.show()
```

```python
# spliting dataset as training and testing

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
y_train
```

# model fitting

```python
# define model

from sklearn.linear_model import LinearRegression
reg =  LinearRegression()
reg.fit(x_train,y_train)
```

```python
# predict

res=reg.predict([[3]])
res
```

```python
res=reg.predict(x_test)
res
```

```python
y_test
```

```python
# visualize 

plt.scatter(x,y)
res=reg.predict(x)
plt.plot(x,res,color='red')
plt.show()
```

```python
# for user input

years = int(input('enter year of experience:'))
myin=pd.DataFrame(columns=['YearsExperience'],data=[[years]])
res=reg.predict(myin)
print('ur salary must be close to:',round(res[0,0]))
```
