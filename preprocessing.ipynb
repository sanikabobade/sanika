```python
# handling missing values(mean and scikit learn)

from sklearn.impute import SimpleImputer
si = SimpleImputer (missing_values = np.nan , strategy = 'mean')
ds[['Age','Salary']] =  si.fit_transform(ds[['Age','Salary']])
```

```python
ds
```

```python
ds['Country'].unique()
```

```python
# encoding categorical coiumn (one-hot encoder)

from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse_output = False)
encoded_data = ohe.fit_transform(ds[['Country']])
encoded_data
```

```python
# make column headers and dataframe for aaray

cm = ohe.get_feature_names_out(['Country']) #it takes column names
encoded_df = pd.DataFrame(encoded_data , columns=cm)
encoded_df
```

```python
ds
```

```python
# joining original dataset and new dataframe

ds=pd.concat([ds,encoded_df],axis=1)
ds
```

```python
# droping categorical column

ds=ds.drop(columns = ['Country'])
```

```python
# label encoding

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
ds['Purchased']=le.fit_transform(ds['Purchased'])
```

```python
ds
```

```python
# feature scaling (reduce larger data values to its smaller data values)

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
ds[['Age', 'Salary']] = sc.fit_transform(ds[['Age', 'Salary']])
```

```python
ds
```

```python
# divide dataset as independant(input) ad dependant(output) variable-Spliting independant and dependant

x=ds[['Age','Salary','Country_France','Country_Germany','Country_Spain']]
y=ds[['Purchased']]
x
y
```

```python
# spliting dataset as training and testing

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```
